## <center>Тема проекта: "Прогноз отказа клиента от услуг компании, предоставляющей телекоммуникационные услуги" </center>

---

## 1. Описание проекта

   Данный учебный проект направлен на определение оптимальной модели машинного обучения, которая позволит спрогнозировать
   возможный ухода клиента из компании. Модель решает задачу бинарной классификации и на выходе должна предсказывать откажется ли
   клиент от услуг компании или продолжит ими пользоваться.

Метки классов:

*  **1** - Клиент компании относится к группе риска;
*  **0** - Клиент компании относится к лояльным клиентам.

---

# 2. Бизнес-цель проекта

Реализовать автоматизированную технологию анализа поведения клиентов
телекоммуникационной компании, которая позволит бизнесу динамически
регулировать свою политику по отношению к реалиям рынка.

---

# 3. Основные задачи проекта

## Этап 1. Загрузка данных и ETL

Датасет был загружен с платформы Kaggle
Ссылка на страницу: *"https://www.kaggle.com/datasets/blastchar/telco-customer-churn"*

**Результаты**:
- 3 признака являются непрерывными;
- 6 признаков являются бинарными;
- 10 признаков являются мультиклассовыми;
- Колонка tenure содержит нулевые значения, которые соответствует новым пользователям, не имеющих ещё истории оплаты (TotalCharges);
- Были заполнены пропуски в колонке TotalCharges нулями для новых клиентов;
- Бинарные и мультиклассовые признаки были преобразованы из категориального типа в целочисленный с помощью инструментов LabelEncoder() и pd.get_dummies() соответственно.

---

### Этап 2. EDA и подготовка данных
**Результаты**:
- Баланс классов: целевого класса примерно в 3 раза меньше вторичного.
- Рассмотрена взаимосвязь между непрерывными признаками и целевой переменной: по распределению целевого признака видно, что классы тяжело различимы;
- Выбросы отсутствуют;
- Нелинейность в данных преобладает (низкий коэффициент корреляции между признаками и метками классов)
- Тренд в распределении данных отсутствует
- В результате анализа вклада каждого из признаков в целевой класс были определены наименее информативные признаки и удалены из набора данных

---

### Этап 3. Определение наиболее оптимального метода ИИ для решения задачи бинарной классификации
- Были рассмотрены следующие методы ИИ:
    - Логистическая регрессия (Logistic Regression);
    - Деревья решений (DecisionTreeClf);
    - Случайный лес (RandomForestClf);
    - Метод опорных векторов (SupportVectorClf);
    - Градиентный бустинг (GradientBoostingClf);
    - Категориальный бустинг (CatBoost);
    - Нейросеть.

Метриками для сравнения служили Recall и ROC AUC, рассчитываемые кросс-валидацией. В силу наличия дисбаланса классов
использовалась кросс-валидация с равномерным распределением классов (StratifiedKFold) при делении данных на тренировочную и тестовую выборки.

В результате первичного сравнения моделей по метрикам Recall и ROC-AUC наилучшими оказались (по убыванию):
1) Логистическая регрессия;
2) Метод опорных векторов;
3) CatBoost;

Далее, после обучения вышеперечисленных моделей более детально, в силу того, что показания метрик были схожи, то было принято решение использовать самый быстрый из них - логистическую регрессию.

---

### Этап 4. Создание полноценного пайплайна под выбранную модель машинного обучения на основе ООП
*Описание:* Согласно структуре проекта был разработан код для автоматизации обучения, т.е. создан полноценный пайплайн на основе логистической регрессии, который на входе принимает сырые данные, а на выходе - предсказывает целевой класс.

- **preprocessing.py**:
    - Загрузка данных
    - Предобработка данных
    - Масштабирование числовых признаков
    - Задание бинарных меток классов для категориальных признаков
    - Деление данных на две подборки: тренировочную и тестовую


- **train.py**:
    - Загрузка тренировочных данных
    - Использование оптимального метода по результатам детального анализа методов машинного обучения
    - Сохранение полученной модели

- **predict.py**:
    - Загрузка обученной модели на тренировочных данных
    - Предсказание классов на тестовом наборе данных

- **evaluate.py**:
    -  Precision
    -  Recall
    -  F1-score
    -  ROC-AUC
    -  PR-AUC
    - Model coefficients
    - SHAP

### Этап 5. Интеграция полученной модели в пользовательский интерфейс

**Tkinter** & **Flask**

В реализации пользовательского интерфейса с помощью библиотек tkinter и Flask у пользователя есть возможность
загрузки как предобработанных данных так и необработанных с указанием номера строки, по которому необходимо
осуществить прогноз. Программа выводит на экран прогноз состояния клиента: будет ли он дальше пользоваться услугами компании
или нет с указанием вероятности.
